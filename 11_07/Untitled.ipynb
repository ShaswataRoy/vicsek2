```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')

theta = np.loadtxt('angles.txt',delimiter='\n',unpack=True)
```

```python
np.mean(theta)
plt.hist((theta-np.mean(theta))%(2*np.pi),bins=50);
plt.xlabel('$\\theta$')
plt.ylabel('P($\\theta$)')
```
