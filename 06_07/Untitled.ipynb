```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')
#Extract Data

x,y = np.loadtxt('vicsek40.txt',delimiter='\t',unpack=True)
x1,y1 = np.loadtxt('vicsek100.txt',delimiter='\t',unpack=True)
x2,y2 = np.loadtxt('vicsek1000.txt',delimiter='\t',unpack=True)


xn,yn = np.loadtxt('vicsek40n.txt',delimiter='\t',unpack=True)
x1n,y1n = np.loadtxt('vicsek100n.txt',delimiter='\t',unpack=True)
x2n,y2n = np.loadtxt('vicsek1000n.txt',delimiter='\t',unpack=True)


plt.plot(x,y,label='N=40 polar')
plt.plot(x1,y1,label='N=100 polar')
plt.plot(x2,y2,label='N=1000 polar')
plt.plot(xn,yn,label='N=40 nematic')
plt.plot(x1n,y1n,label='N=100 nematic')
plt.plot(x2n,y2n,label='N=1000 nematic')

plt.xlabel('$\\eta$')
plt.ylabel('Order Parameter')

plt.legend()
plt.xlim(0,3)
```
