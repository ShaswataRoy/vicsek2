```python
time = 0
tau = []
cond = []
for j in range(500):
    cond = []
    time = 0
    for i in range(0,149998):
        if up[i][j]>4.5 or up[i][j]<2.0:
            cond.append(1)
        else:
            cond.append(0)
        
    for i in range(1,149998):
        if cond[i]==1 and cond[i-1]==0:
            tau.append(i-time)
            time = i        
```
